```python
#make it look like a percent
# create a new function called makepercent
def makepercent(x):
    # that, if x is a string, so Country does not get affected,
    if type(x) is str:
        # just returns it, no change
        return x
    # but, if not, return it multiplied by 100, therefore a percent instead of decimal value
    elif x:
        return 100 * x
    # end the function
    else:
        return
    #apply the function to roundedGDP
roundedGDP.applymap(makepercent)
```
